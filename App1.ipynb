```python
df = pd.read_csv("coviddata_clean.csv")
df.head()
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
df['year'] = df['publish_time'].dt.year
```

```python
df['year'].describe()
```

```python
df['abstract_length'] = df['abstract'].astype(str).apply(len)
df['abstract_length'].describe()
```

```python
df['author_count'] = df['authors'].apply(lambda x: len(x.split(',')))
df['author_count'].unique()
```

```python
# Set author_count to 0 where 'authors' contain "unknown"
df.loc[df['authors'].str.lower().str.contains('unknown', na=False), 'author_count'] = 0
```

```python
# Count the number of papers published per year
# Sort_index ensures the years are in chronological order

year_counts = df['year'].value_counts().sort_index()

# Display results 
year_counts[year_counts > 100]
print(max(year_counts))
```

```python
#lets identify the top journal
journal_count = df["journal"].value_counts()
# Display the top 10 journals
top_journals = journal_count.head(15)
print(top_journals)
```

```python
# Combine all titles into one string (remove missing ones)
all_titles = ' '.join(df['title'].dropna().astype(str))

# Convert to lowercase
all_titles = all_titles.lower()

# Remove punctuation manually
for ch in ['.', ',', '!', '?', ':', ';', '-', '(', ')', '"', "'"]:
    all_titles = all_titles.replace(ch, '')

# Split into words
words = all_titles.split()

# Count word frequency manually
word_counts = {}
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Sort by frequency (highest first)
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Display the top 20 words
for word, count in sorted_words[:30]:
    print(f"{word}: {count}")

```

visualizing publicatiobns per year

```python
# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-', linewidth=2)
plt.title('Number of COVID-19 Publications Over Time')
plt.xlabel('Publication Year')
plt.ylabel('Number of Papers')
plt.grid(True)
plt.show()
```

Barchart for top publishing journals

```python
plt.figure(figsize=(10, 6))
plt.barh(top_journals.index, top_journals.values, color='skyblue')
plt.title('Top 10 Journals Publishing COVID-19 Research')
plt.xlabel('Number of Publications')
plt.ylabel('Journal')
plt.gca().invert_yaxis()  # Highest bar at the top
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud


# Combine all titles into one large string
all_titles = " ".join(df['title'].dropna().astype(str))

# Create the word cloud
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      max_words=100,
                      colormap='viridis').generate(all_titles)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Frequent Words in Paper Titles", fontsize=14)
plt.show()
```

```python


# Count the number of papers per source
source_counts = df['source_x'].value_counts()

# Create the plot
plt.figure(figsize=(10, 6))
bars = plt.bar(source_counts.index, source_counts.values, color='teal')

# Add titles and labels
plt.title('Distribution of Paper Counts by Source', fontsize=14)
plt.xlabel('Source', fontsize=12)
plt.ylabel('Number of Papers', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Add exact counts on each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x position (center of bar)
        yval + 5,                           # y position (slightly above bar)
        int(yval),                          # label text (integer count)
        ha='center', va='bottom', fontsize=10
    )

# Adjust layout
plt.tight_layout()
plt.show()
```
